```python
import pandas as pd

df = pd.read_csv("forestfires.csv")

df.describe()
```

```python
import matplotlib.pyplot as plt

#plt.hist(df.area)
k = {}

k['jan'] = 0
k['feb'] = 0
k['mar'] = 0
k['apr'] = 0
k['may'] = 0
k['jun'] = 0
k['jul'] = 0
k['aug'] = 0
k['sep'] = 0
k['oct'] = 0
k['nov'] = 0
k['dec'] = 0
  
for month in df.month:
    k[month] += 1
    
plt.bar([*k], k.values())
plt.show()
```
